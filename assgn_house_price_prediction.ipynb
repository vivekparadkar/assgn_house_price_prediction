### Categorical Variables

1. MSSubClass: Identifies the type of dwelling involved in the sale.
2. MSZoning: Identifies the general zoning classification of the sale.
3. Street: Type of road access to property
4. Alley: Type of alley access to property
5. LotShape: General shape of property
6. LandContour: Flatness of the property
7. Utilities: Type of utilities available
8. LotConfig: Lot configuration
9. LandSlope: Slope of property
10. Neighborhood: Physical locations within Ames city limits
11. Condition1: Proximity to various conditions
12. Condition2: Proximity to various conditions (if more than one is present)
13. BldgType: Type of dwelling
14. HouseStyle: Style of dwelling
15. OverallQual: Rates the overall material and finish of the house
16. OverallCond: Rates the overall condition of the house
17. RoofStyle: Type of roof
18. RoofMatl: Roof material
19. Exterior1st: Exterior covering on house
20. Exterior2nd: Exterior covering on house (if more than one material)
21. MasVnrType: Masonry veneer type
22. ExterQual: Evaluates the quality of the material on the exterior
23. ExterCond: Evaluates the present condition of the material on the exterior
24. Foundation: Type of foundation
25. BsmtQual: Evaluates the height of the basement
26. BsmtCond: Evaluates the general condition of the basement
27. BsmtExposure: Refers to walkout or garden level walls
28. BsmtFinType1: Rating of basement finished area
29. BsmtFinType2: Rating of basement finished area (if multiple types)
30. Heating: Type of heating
31. HeatingQC: Heating quality and condition
32. CentralAir: Central air conditioning
33. Electrical: Electrical system
34. KitchenQual: Kitchen quality
35. Functional: Home functionality (Assume typical unless deductions are warranted)
36. FireplaceQu: Fireplace quality
37. GarageType: Garage location
38. GarageFinish: Interior finish of the garage
39. GarageQual: Garage quality
40. GarageCond: Garage condition
41. PavedDrive: Paved driveway
42. PoolQC: Pool quality
43. Fence: Fence quality
44. MiscFeature: Miscellaneous feature not covered in other categories
45. SaleType: Type of sale
46. SaleCondition: Condition of sale
